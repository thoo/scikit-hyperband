```python
import pandas as pd
import lightgbm as lgb
from sklearn.datasets import load_breast_cancer
from hyperband import HyperbandSearchCV
```

```python
# import data
X,y = load_breast_cancer(return_X_y=True)
```

```python
# Set paramters for the models
band_params = {'boosting_type': 'gbdt',
 'class_weight': None,
 'colsample_bytree': 0.9,
 'importance_type': 'split',
 'learning_rate': 0.01,
 'max_depth': 20,
 'min_child_samples': 25,
 'min_child_weight': 0,
 'min_split_gain': 0,
 'n_estimators': 4000,
 'n_jobs': -1,
 'num_leaves': 200,
 'objective': 'binary',
 'random_state': None,
 'reg_alpha': 0,
 'reg_lambda': 0,
 'silent': True,
 'subsample': 0.8,
 'subsample_for_bin': 200000,
 'subsample_freq': 1,
 'metric': 'auc',
 'max_bin': 100,
 'verbose': -1,
 'scale_pos_weight': 1}
# Set the tuning parameters for the model
param_dict = {'max_depth':[2,5]}
```

```python
search = HyperbandSearchCV(lgb.LGBMClassifier(**band_params), param_dict, 
                           resource_param='n_estimators',verbose=100,max_iter=100,min_iter=1,
                           scoring='roc_auc')
```

```python
fit_params={'early_stopping_rounds':10,'eval_set':[(X,y)]}
```

```python
search.fit(X,y,**fitt_params)
```
